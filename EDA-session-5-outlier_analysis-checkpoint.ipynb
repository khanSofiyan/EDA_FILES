**Finding the Out outliers**

- we already know that outlier aviable less than Q1-1.5*IQR

- Step-1:

    - Calculate Q1=25p

    - Calculate Q2=50p

    - Calculate Q3=75p

- Step-2:

    - Calculate IQR = Q3-Q1

- Step-3:

    - Calculate lb = Q1-1.5*IQR

    - Calcultate ub = Q3+1.5*IQR 

- Step-4:

    - con1= wage_data <lb

    - con2= wage_data >up

    - con3= con1 or con2 / con1 and con2

- Step-5:

    - data[cont3]

```python
Q1=np.percentile(wage_data,25)
Q2=np.percentile(wage_data,50)
Q3=np.percentile(wage_data,75)
#Step-2:
IQR=Q3-Q1
#step-3:
lb= Q1-1.5*IQR
ub= Q3+1.5*IQR
#Step-4:
Con1= wage_data<lb
Con2= wage_data>ub
Con3= Con1|Con2
#bitwise operators used to campare
#Bits data==== 0 and 1===== True false
#step-5:
outliers_data=wage_data[Con3]
outliers_data
```

**How to deal outliers**

- 1. We can drop th outliers

    - Droping outliers is an easy task , but no recommended

    - If you drop outliers we are droping the row

    - which means we are droping the other column  values also

    - May be some import information we might loose from other columns values

    - But in industry if outliers percentage less than 2% of the total data, we can drop

    - The number of outliers in pawage is 427

    - The number of total observation are 25480

    - The percentage is 1.67 which is less than 2%

- 2. fill with Medain value

    - We already know that outliers dose not affect by medain

    - so it is the best practice to fill the outliers with median

- 3. capping or florring

    - Instead of fill with median

    - we can fill with certian values

    - the outliers which are less than LB fill with LB values

    - the outliers which are greater than UB fill with UB values
