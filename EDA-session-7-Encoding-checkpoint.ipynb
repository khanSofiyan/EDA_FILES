**Map**

- map is one method to convert categorical values to numerical 

- Take one categorical column

- Get the unique lables first 

- make a dictionary with assign a number to each lable

- For example case_status has two lables

    - Certified 

    - Denied 

- Assign 0 to certified , 1 to Denied

- Create a dictionary lables as keys , number as values only

- {'Certified':0,'Denied':1}

```python
# step-1: read the columns
# step-2: get the unque lables
# step-3: make a dictionary
# step-4: apply the map
```

```python
# i want to apply the map method for all categorical columns
# step:1 - we need to get unique lables
# step:2 - we need to get a number = len unique lables
# step:3 - we need to create a dictionary 
```

```python
unique_lables=sorted(visa_df['case_status'].unique())
list1=[i for i in range(len(unique_lables))]
dict1=dict(zip(unique_lables,list1))
dict1
visa_df['case_status']=visa_df['case_status'].map(dict1)
visa_df
```

```python
visa_df=pd.read_csv('Visadataset.csv')
cat_columns=visa_df.select_dtypes(include="object").columns
for i in cat_columns:
    unique_lables=sorted(visa_df[i].unique())
    list1=[i for i in range(len(unique_lables))]
    dict1=dict(zip(unique_lables,list1))
    dict1
    visa_df[i]=visa_df[i].map(dict1)
visa_df
```

**np.where**

- using np.where also we can change the cartegorical data to Numerical

- But np.where is only for binary condition

- if condition is true : will apply true value

- if condition is false : will apply false value

- so np.where use for only binary lables

- A column should have only two unique lables

```python
# value =='certified'  replace with 0
# otherwise 1
visa_df=pd.read_csv('Visadataset.csv')
con=visa_df['case_status']=='Certified'
visa_df['case_status']=np.where(con,0,1)
visa_df
```

**one Hot encoder**

- On represent with 1 
- off represent with 0

    - Denied 
- one hot encoder new column which is equal to number of unique lables
- For example case status will create two new extra columns

    - case_status_Denied
