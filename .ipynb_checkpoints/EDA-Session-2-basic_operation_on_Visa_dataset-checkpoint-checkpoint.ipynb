$Take$

- first argument : indices

- indices will take some values in array : in a list

- It tries to extract the data

- the values we provides represents column  or rows??

- Axis = 0 represent about rows

- Axis = 1 represent about columns

- by default axis mentioned as zero

```python
visa_df.take([100,200,300])
# its retrive 3 rows wohich having index 100,200,300
```

```python
visa_df.take([100,200,300],axis=1)
# how this will check for 100th columns 20th columns and 30th column
# but originally we have only 12 column (0,11)
# will get error
```

```python
#case-3 : select 3 columns 2nd columns , 5th and 7th columns
#using take
visa_df.take([2,5,7],axis=1)

# the output also behave as dataframe
```

```python
#case-4 : slicing will fail
visa_df.take([2:5])
```
